```python
from konlpy.tag import Okt # 명사 추출
from collections import Counter # 빈도계산
import csv
```

```python
filename = './저상버스_부족.csv'
f = open(filename, 'r', encoding='utf-8')
news = f.read()
```

```python
# Okt 객체 생성
okt = Okt()
noun = okt.nouns(news)
for i, v in enumerate(noun):
    if len(v) < 2:
        noun.pop(i)
        
count  = Counter(noun)
f.close()
```

```python
# 명사 빈도 카운트
noun_list = count.most_common(180)
for v  in noun_list:
    print(v)
```
