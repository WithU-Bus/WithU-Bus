```python
import pandas as pd
```

# 부산시 저상버스 현황

```python
# 저상버스 운행노선수 현황

df = pd.read_csv('./data/한국교통안전공단_지역별 시내버스의 저상버스 운행 노선수 현황_20220830.csv',encoding='utf-8')
df
```

```python
# 저상버스 보급현황

df1 = pd.read_csv('./data/한국교통안전공단_지역별 저상버스 보급현황_20211231.csv',encoding='utf-8')
df1
```

```python
# 저상버스 노선수 현황과 보급현황 합치기

lowbus_info = pd.merge(df,df1)
lowbus_info
```

```python
# 부산시 저상버스 현황
busanlowbus = lowbus_info[lowbus_info['구분']=='부산광역시']
busanlowbus
```

```python
df2 = pd.read_csv('./res/busnumber.csv')
df2
```

```python
busanlowbus['전체 버스 대수'] = df2['평일_운행대수'].loc[183]
busanlowbus
```

```python
busanlowbus.rename(columns={"시내버스(전체) 운행노선수":"시내버스 운행노선수","시내버스(저상버스) 운행노선수":"저상버스 운행노선수","2021년 저상버스 대수":"저상버스 대수","2021년 보급률":"저상버스 보급률","전체 버스 대수":"시내버스 대수"},inplace=True)
busanlowbus
```

```python
busanlowbus = busanlowbus[['구분','시내버스 운행노선수','저상버스 운행노선수','저상버스 운행노선 비율','시내버스 대수','저상버스 대수','저상버스 보급률']]
busanlowbus
```

```python
busanlowbus.to_csv('./res/busanlowbus.csv')
```
