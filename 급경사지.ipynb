### 급경사지 데이터 불러오기(위험등급X)

```python
youngdo = pd.read_excel('data\영도구_급경사지.xlsx')
youngdo
```

```python
gumjung = pd.read_excel('data\금정구_급경사지.xlsx')
gumjung
```

```python
b = pd.concat([youngdo, gumjung])
b = b.reset_index(drop='true')
b
```

### 위험등급O 시각화

```python
map3_danger = folium.Map((35.1795543,129.0756416), zoom_start=11)

for i in a.index:
  sub_lat = a.loc[i, 'X']
  sub_long = a.loc[i, 'Y']
  
  # title = f"{a.loc[i, '관리구']} - {a.loc[i, '위험등급']}"
  
  if a.loc[i, '위험등급'] == 'B': 
    color='green',
    title = f"{a.loc[i, '관리구']} - {a.loc[i, '위험등급']}"
  elif a.loc[i, '위험등급'] == 'C':
    color='yellow'
    title = f"{a.loc[i, '관리구']} - {a.loc[i, '위험등급']}"
  elif a.loc[i, '위험등급'] == 'D':
    color='red'
    title = f"{a.loc[i, '관리구']} - {a.loc[i, '위험등급']}"
  
  folium.CircleMarker(location=[sub_lat, sub_long],
                      radius=3,
                      color=color,
                      tooltip=title).add_to(map3_danger)

map3_danger

```

### 위험등급(X) 시각화

```python
map = folium.Map((35.1795543,129.0756416), zoom_start=11)

for i in b.index:
  sub_lat = b.loc[i, 'X']
  sub_long = b.loc[i, 'Y']
  
  title = f"{b.loc[i, '관리구']}"
  
  color='blue'
  
  folium.CircleMarker(location=[sub_lat, sub_long],
                      radius=3,
                      color=color,
                      tooltip=title).add_to(map)

map
```
