```python
# csv 불러오기
bs = pd.read_csv('./res/부산버스정보시스템.csv', encoding='cp949')
```

#### 부산 정보 안내기관

```python
# csv 불러오기
bi = pd.read_csv('./res/부산버스정보안내기현황.csv', encoding='cp949')
```

```python
bi
```

```python
# 컬럼값 변경
bs['정류장 구분'] = '시내'
```

```python
bs
```

```python
# 필요없는 컬럼 삭제
bi = bi.drop(['설치년도','주소','데이터기준일자'], axis=1)
```

```python
bi
```

```python
# 컬럼 수 맞추기
bi['정류소 아이디'] = np.NaN
bi['정류장 구분'] = np.NaN
```

```python
bi.head()
```

```python
bs.head()
```

```python
# 컬럼 명 바꾸기
bi.rename(columns = {'정류장번호' : '정류소 번호','위치명(정류장명)' : '정류소 명', '좌표정보(위도)':'위도',
                     '좌표정보(경도)':'경도'}, inplace=True)
bi
```

```python
# 컬럼 명 바꾸기
bs.rename(columns={'GPS X 좌표' : '경도', 'GPS Y 좌표': '위도'}, inplace=True)
bs
```

```python
# 컬럼 위치 바꾸기
bs = bs[['정류소 번호', '정류소 아이디', '정류소 명', '위도', '경도', '정류장 구분']]
bs
```

```python
# 컬럼 확인
bi.columns
```

```python
# 컬럼 위치 동일하게 바꾸기
bi = bi[['정류소 번호', '정류소 아이디', '정류소 명', '위도', '경도', '정류장 구분']]
bi
```

```python
# 두 데이터프레임 아래로 합치기
bus = pd.concat([bs, bi], axis=0)
bus
```

```python
# 중복되는 컬럼 찾기
bus.duplicated('정류소 명')
```

```python
# 정류소명 컬럼 기준으로 중복 제거
bus2 = bus.drop_duplicates(['정류소 명'], keep='first')
bus2
```

```python
bus2 = bus2.reset_index(drop=True)
bus2
```

```python
bus2.to_csv('./res/busstation.csv', encoding='cp949')
```
