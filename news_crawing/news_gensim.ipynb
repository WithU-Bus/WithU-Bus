#### 라이브러리

```python
import pandas as pd
import os
from gensim.summarization.summarizer import summarize
```

#### 데이터 전처리

```python
# 파일에 있는 모든 csv 파일 불러오기

# 불러온 파일이 있는 경로
path = './data/저상버스_부족'
file_list = os.listdir(path)
file_list_py = [file for file in file_list if file.endswith('.csv')] # 파일명이 끝이 .csv 인 경우
```

```python
# csv 파일들을 DataFrame으로 불러와서 concat

news = pd.DataFrame()
for i in file_list_py:
    data = pd.read_csv(path+'/'+i, encoding='utf-8')
    news = pd.concat([news, data])
    
    news = news.reset_index(drop=True)
news
```

```python
news = news.drop(columns=['Unnamed: 0','날짜','기사제목'], axis=0)
# summarize를 사용하기 위해 오류가 나는 문장을 고쳐 줌
news["기사내용"] = news["기사내용"].str.replace("\t", "").str.replace(".", "\n")
news
```

```python
# 리스트로 담기
sum_list = []

for i in range(0, len(news)):
    text = news['기사내용'][i]
    text_list = (''.join(summarize(text, word_count=2000).split('\n')))
    
    sum_list.append(text_list)
    
```

```python
total = pd.DataFrame(sum_list, columns=['기사 요약'])
total
```

```python
total.to_csv('./res/저상버스_부족_요약본.csv')
```
