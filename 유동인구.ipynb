```python
import pandas as pd
```

# 구별 유동인구

```python
Gangseo = pd.read_csv('./data2/부산광역시 강서구 유동인구 수.csv')
Geumjeong = pd.read_csv('./data2/부산광역시 금정구 유동인구 수.csv')
Gijang = pd.read_csv('./data2/부산광역시 기장군 유동인구 수.csv')
Nam = pd.read_csv('./data2/부산광역시 남구 유동인구 수.csv')
Dong = pd.read_csv('./data2/부산광역시 동구 유동인구 수.csv')
Dongnae = pd.read_csv('./data2/부산광역시 동래구 유동인구 수.csv')
jin = pd.read_csv('./data2/부산광역시 부산진구 유동인구 수.csv')
Buk = pd.read_csv('./data2/부산광역시 북구 유동인구 수.csv')
Sasang = pd.read_csv('./data2/부산광역시 사상구 유동인구 수.csv')
Saha = pd.read_csv('./data2/부산광역시 사하구 유동인구 수.csv')
seo = pd.read_csv('./data2/부산광역시 서구 유동인구 수.csv')
suyeong = pd.read_csv('./data2/부산광역시 수영구 유동인구 수.csv')
yeonje = pd.read_csv('./data2/부산광역시 연제구 유동인구 수.csv')
yeongdo = pd.read_csv('./data2/부산광역시 영도구 유동인구 수.csv')
jung = pd.read_csv('./data2/부산광역시 중구 유동인구 수.csv')
haeundae = pd.read_csv('./data2/부산광역시 해운대구 유동인구 수.csv')
Gangseo.head()
```

```python
population = pd.concat([Gangseo, Geumjeong, Gijang, Nam, Dong, Dongnae, jin, Buk, Sasang, Saha, seo, suyeong, yeonje, yeongdo, jung, haeundae, Gangseo])
population.head()
```

```python
population.tail()
```

```python
population = population.drop(['INDEX_KEY', 'CTPV_CD', 'CTPV_NM', 'SGG_CD', 'SEX_DV', 'AGRDE_CLS', 'REVISN_AMBLT_PUL_CNT', 'REVISN_NMBLT_PUL_CNT'], axis=1)
population
```

```python
population = population.rename(columns={'CRTR_YMD' : '기준일자', 'CRTR_WEEK' : '기준주', 'DWK_NM' : '요일명', 'SGG_NM' : '시군구명', 'TMZN_CD' : '시간대코드'})
population
```

```python
population.to_csv('./res/population.csv', index=False)
```
