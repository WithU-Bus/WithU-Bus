```python
import pandas as pd
from collections import Counter
from konlpy.tag import Okt
import csv
```

# 부산_저상버스

```python
filename = './res/news_summary/부산_저상버스_요약_0.4.csv'
f = open(filename, 'r', encoding='utf-8')
news = f.read()
```

```python
# Okt 객체 생성
okt = Okt()
noun = okt.nouns(news)
for i, v in enumerate(noun):
    if len(v) < 2:
        noun.pop(i)
        
count  = Counter(noun)
f.close()
```

```python
# 명사 빈도 카운트
noun_list = count.most_common(180)
for v  in noun_list:
    print(v)
```

```python
# csv 파일로 저장
with open("./res/wordcloud/부산_저상버스_빈도.csv","w", newline='') as f:
    csvw = csv.writer(f)
    for v in noun_list:
        csvw.writerow(v)
```

```python
total_all = pd.read_excel('./res/wordcloud/빈도Total.xlsx', index_col=0)
total_all
```

```python
total_all = total_all.groupby(['단어'])['빈도수'].sum()
total_all
```

```python
from wordcloud import WordCloud
import matplotlib.pylab as plt
from PIL import Image
import pandas as pd
import numpy as np
```

```python
wc = WordCloud(font_path='malgun', background_color='#f5f5f5', width=1280, height=760)
cloud = wc.generate_from_frequencies(total_all)

plt.figure(figsize=(30,10))
plt.imshow(cloud)
plt.axis('off')
plt.show()
```

```python
img = Image.open('res/wordcloud/bus.png')
img_array = np.array(img)

wc = WordCloud(font_path='malgun', width=400, height=200, scale=2.0, max_font_size=500, mask=img_array, background_color = 'white')
gen = wc.generate_from_frequencies(total_all)

plt.figure(figsize=(15,10))
plt.axis(False)
plt.imshow(gen)
```
