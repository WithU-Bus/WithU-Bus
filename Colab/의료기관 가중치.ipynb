# 버스정류장

```python
bus_stop = pd.read_csv('/content/drive/MyDrive/데이터/bus_stop.csv', encoding='utf-8', index_col=0)
bus_stop
```

```python
bus_stop = bus_stop.drop_duplicates(['X', 'Y'])
bus_stop
```

# 의료기관

```python
infra = pd.read_csv('/content/drive/MyDrive/데이터/infra.csv')
infra
```

```python
infra_재활 = infra[infra['종류'].str.contains('재활')]
infra_병원 = infra[infra['종류'].str.contains('병원')]
infra_의료 = pd.concat([infra_재활, infra_병원])
infra_의료 = infra_의료.reset_index(drop=True)
infra_의료
```

```python
infra_의료 = infra_의료[['시설명', '위도', '경도']]
infra_의료
```

```python
의료_dict = infra_의료.set_index('시설명').T.to_dict('list')
의료_dict
```

# 거리계산

```python
from haversine import haversine
```

```python
import tensorflow as tf
```

```python
with tf.device('/device:GPU:0'):
  df_all = pd.DataFrame(columns=["base_name", "bus_name", "거리차이", "score"])

  for i in range(0, len(의료_dict)) :
      base_nm = list(의료_dict.keys())[i]
      base_xy = list(의료_dict.items())[i][1]

      for j in range(0, len(bus_stop_dict)) :
          bus_nm = list(bus_stop_dict.keys())[j]
          bus_xy = list(bus_stop_dict.items())[j][1]

          ## 최종 거리 계산 결과값
          result = haversine(base_xy, bus_xy, unit='m')

          score = 0
          if result <= 50:
              score = 5
          elif 50 < result <= 75:
              score = 4
          elif 75 < result <= 100:
              score = 3
          elif 100 < result <= 125:
              score = 2
          elif 125 < result <= 150:
              score = 1
          df_all = pd.concat([df_all, 
                            pd.DataFrame({'base_name':[base_nm],'bus_name':[bus_nm], '거리차이':[result], 'score':[score]})], ignore_index=True)
  df_all
```
