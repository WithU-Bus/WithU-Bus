```python
bus_stop = pd.read_csv('/content/drive/MyDrive/데이터/bus_stop.csv', encoding='utf-8', index_col=0)
bus_stop
```

```python
bus_stop = bus_stop.drop_duplicates(['X', 'Y'])
bus_stop
```

```python
bus_stop = bus_stop[['NAME', 'BUSSTOPID', 'Y', 'X']]
bus_stop
```

```python
bus_stop_dict = bus_stop.set_index(['NAME', 'BUSSTOPID']).T.to_dict('list')
bus_stop_dict
```

# 직장

```python
employ = pd.read_excel('/content/drive/MyDrive/데이터/장애인일자리_공공기관.xlsx')
employ
```

```python
employ = employ[['기관명', '위도', '경도']]
employ
```

```python
employ_dict = employ.set_index('기관명').T.to_dict('list')
employ_dict
```

# 거리계산

```python
from haversine import haversine
```

```python
import tensorflow as tf
```

```python
with tf.device('/device:GPU:0'):
  df_all = pd.DataFrame(columns=["base_name", "bus_name", "거리차이", "score"])

  for i in range(0, len(employ_dict)) :
      base_nm = list(employ_dict.keys())[i]
      base_xy = list(employ_dict.items())[i][1]

      for j in range(0, len(bus_stop_dict)) :
          bus_nm = list(bus_stop_dict.keys())[j]
          bus_xy = list(bus_stop_dict.items())[j][1]

          ## 최종 거리 계산 결과값
          result = haversine(base_xy, bus_xy, unit='m')

          score = 0
          if result <= 50:
              score = 5
          elif 50 < result <= 75:
              score = 4
          elif 75 < result <= 100:
              score = 3
          elif 100 < result <= 125:
              score = 2
          elif 125 < result <= 150:
              score = 1
          df_all = pd.concat([df_all, 
                            pd.DataFrame({'base_name':[base_nm],'bus_name':[bus_nm], '거리차이':[result], 'score':[score]})], ignore_index=True)
  df_all
```
