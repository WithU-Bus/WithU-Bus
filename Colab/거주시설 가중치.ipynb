# 거주시설

```python
infra = pd.read_csv('/content/drive/MyDrive/데이터/infra.csv')
infra
```

```python
infra_a = infra[infra['종류'].str.contains('장애인주간보호센터')]
infra_b = infra[infra['종류'].str.contains('거주')]
```

```python
infra_거주 = pd.concat([infra_a, infra_b])
infra_거주 = infra_거주.reset_index(drop=True)
infra_거주
```

```python
infra_거주 = infra_거주[['시설명', '위도', '경도']]
infra_거주 = infra_거주.reset_index(drop=True)
infra_거주
```

```python
거주_dict = infra_거주.set_index('시설명').T.to_dict('list')
거주_dict
```

```python
from haversine import haversine
```

```python
import tensorflow as tf
```

```python
with tf.device('/device:GPU:0'):
  df_all = pd.DataFrame(columns=["base_name", "bus_name", "거리차이", "score"])

  for i in range(0, len(거주_dict)) :
      base_nm = list(거주_dict.keys())[i]
      base_xy = list(거주_dict.items())[i][1]

      for j in range(0, len(bus_stop_dict)) :
          bus_nm = list(bus_stop_dict.keys())[j]
          bus_xy = list(bus_stop_dict.items())[j][1]

          ## 최종 거리 계산 결과값
          result = haversine(base_xy, bus_xy, unit='m')

          score = 0
          if result <= 50:
              score = 5
          elif 50 < result <= 75:
              score = 4
          elif 75 < result <= 100:
              score = 3
          elif 100 < result <= 125:
              score = 2
          elif 125 < result <= 150:
              score = 1
          df_all = pd.concat([df_all, 
                            pd.DataFrame({'base_name':[base_nm],'bus_name':[bus_nm], '거리차이':[result], 'score':[score]})], ignore_index=True)
  df_all
```
