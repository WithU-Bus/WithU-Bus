```python
import pandas as pd
```

```python
bus_stop = pd.read_csv('/content/drive/MyDrive/데이터/bus_stop.csv', encoding='utf-8', index_col=0)
bus_stop
```

```python
bus_stop = bus_stop.drop_duplicates(['X', 'Y'])
bus_stop
```

```python
bus_stop = bus_stop[bus_stop['SIGUNNAME'].str.contains('부산')]
bus_stop = bus_stop.reset_index(drop=True)
bus_stop
```

```python
bus_stop = bus_stop[['NAME', 'BUSSTOPID', 'GUNAME']]
bus_stop
```

```python
bus_stop['인구_score'] = 0

for i in range(len(bus_stop)) :
  if bus_stop['GUNAME'][i] == '부산진구':
    bus_stop['인구_score'][i] = 5
  elif bus_stop['GUNAME'][i] == '사하구':
    bus_stop['인구_score'][i] = 4
  elif bus_stop['GUNAME'][i] == '해운대구':
    bus_stop['인구_score'][i] = 3
  elif bus_stop['GUNAME'][i] == '북구':
    bus_stop['인구_score'][i] = 2
  elif bus_stop['GUNAME'][i] == '사상구':
    bus_stop['인구_score'][i] = 1
  else:
    bus_stop['인구_score'][i]=0
bus_stop
```
